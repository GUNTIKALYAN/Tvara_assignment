```python
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
embeddings = HuggingFaceEmbeddings(
    model_name="intfloat/e5-small-v2"
)
```

```python
sentences = [
    """Cricket is one of the most loved sports worldwide.
It is especially popular in countries like India and Australia.
Players compete in formats like Test, ODI, and T20.""",

    """Machine learning enables systems to learn from data automatically.
It helps computers identify patterns without explicit programming.
ML powers recommendations, predictions, and automation.""",

    """Artificial Intelligence is reshaping modern technology.
It combines machine learning, reasoning, and perception.
Industries like healthcare, finance, and education depend on AI.""",

    """Deep learning is a powerful branch of machine learning.
It uses neural networks with multiple layers to learn complex patterns.
Deep learning excels in vision, NLP, and speech tasks.""",

    """Data science extracts insights from structured and unstructured data.
It involves statistics, visualization, and analytical thinking.
Businesses rely on data science for decision-making."""
]
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=120,
    chunk_overlap=20
)
```

```python
chunks = text_splitter.create_documents(sentences)
```

```python
db2 = FAISS.from_documents(chunks,embeddings)
```

```python
query = "tell me about the cricket?"
results = db2.similarity_search(query,k=2)
```

```python
print(query)
for r in results:
    print(r.page_content)
```
